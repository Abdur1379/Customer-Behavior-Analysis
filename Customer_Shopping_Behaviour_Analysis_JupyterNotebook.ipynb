```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename (columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df.columns
```

```python
frequency_mapping = {
    'fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Option 1: If you want to fill NaN with 0 and convert to int
df['purchase_frequency_days'] = df['purchase_frequency_days'].fillna(0).astype(int)
```

```python
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype('Int64')
```

```python
# Replace missing purchase_frequency_days based on frequency_of_purchases
df.loc[(df['frequency_of_purchases'] == 'Fortnightly') & (df['purchase_frequency_days'].isna()), 'purchase_frequency_days'] = 14
df.loc[(df['frequency_of_purchases'] == 'Weekly') & (df['purchase_frequency_days'].isna()), 'purchase_frequency_days'] = 7
df.loc[(df['frequency_of_purchases'] == 'Monthly') & (df['purchase_frequency_days'].isna()), 'purchase_frequency_days'] = 30
df.loc[(df['frequency_of_purchases'] == 'Quarterly') & (df['purchase_frequency_days'].isna()), 'purchase_frequency_days'] = 90
df.loc[(df['frequency_of_purchases'] == 'Annually') & (df['purchase_frequency_days'].isna()), 'purchase_frequency_days'] = 365

# Convert to whole number (keep NaN-compatible type)
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype('Int64')
```
